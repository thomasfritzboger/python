```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
from IPython.display import Image

image = cv2.imread("appollo_11.jpg",cv2.IMREAD_COLOR)
plt.imshow(image[:,:,::-1])
```

```python
image_line = image.copy()

cv2.line(image_line, (75,50), (125,50), (0,255,255), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(image_line[:,:,::-1])
```

```python
# Draw circle
image_circle = image.copy()

cv2.circle(image_circle, (200,140), 100, (0,0,255), thickness=3, lineType=cv2.LINE_AA)
cv2.circle(image_circle, (320,240), 20, (255,255,0), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(image_circle[:,:,::-1])
```
