```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image
%matplotlib inline

cb_img = cv2.imread("checkerboard_18_18.png",0)
print(cb_img)
```

```python
cb_img_copy = cb_img.copy()
cb_img_copy[2,2]=200
cb_img_copy[2,3]=200
cb_img_copy[3,2]=200
cb_img_copy[3,3]=200

plt.imshow(cb_img_copy,cmap='gray')
print(cb_img_copy)
```

```python
img_NZ_bgr = cv2.imread("NZ_pic.jpg", cv2.IMREAD_COLOR)
img_NZ_rgb = img_NZ_bgr[:,:,::-1]
plt.imshow(img_NZ_rgb)
```

```python
cropped_region = img_NZ_rgb[175:350,350:600]
plt.imshow(cropped_region)
```

```python
resized_cropped_region_2x = cv2.resize(cropped_region,None,fx=2,fy=2)
plt.imshow(resized_cropped_region_2x)
```

```python
desired_width = 200
desired_height = 200
dim = (desired_width,desired_height)

# Resize background image to same size as logo image
resized_cropped_region = cv2.resize(cropped_region, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
```

```python
# Method 2: Using 'dsize'
desired_width = 150
aspect_ratio = desired_width/cropped_region.shape[1]
desired_height = int(cropped_region.shape[0]*aspect_ratio)
dim = (desired_width,desired_height)

resized_cropped_region = cv2.resize(cropped_region, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
```

```python
# Flipping images
img_NZ_flip_horz = cv2.flip(img_NZ_rgb,1)
img_NZ_flip_vert = cv2.flip(img_NZ_rgb,0)
img_NZ_flip_both = cv2.flip(img_NZ_rgb,-1)

plt.figure(figsize=[18,5])
plt.subplot(141);plt.imshow(img_NZ_flip_horz);plt.title("Horizontal flip")
plt.subplot(142);plt.imshow(img_NZ_flip_vert);plt.title("Vertictal flip")
plt.subplot(143);plt.imshow(img_NZ_flip_both);plt.title("Both flip")
plt.subplot(144);plt.imshow(img_NZ_rgb);plt.title("Original")
```
