```python
# Basic Image Enhancement Using Mathematical Operations
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
from IPython.display import Image

img_bgr = cv2.imread("NZ_pic.jpg", cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

Image(filename="NZ_pic.jpg")
```

```python
matrix = np.ones(img_rgb.shape, dtype = "uint8") * 50
img_rgb_bright = cv2.add(img_rgb, matrix)
img_rgb_dark = cv2.subtract(img_rgb, matrix)

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_rgb_dark); plt.title("Darker")
plt.subplot(132); plt.imshow(img_rgb); plt.title("Original")
plt.subplot(133); plt.imshow(img_rgb_bright); plt.title("Brighter")
```

```python
matrix1 = np.ones(img_rgb.shape)*.8
matrix2 = np.ones(img_rgb.shape)*1.2

img_rgb_dark = np.uint8(cv2.multiply(np.float64(img_rgb), matrix1))
img_rgb_bright = np.uint8(cv2.multiply(np.float64(img_rgb), matrix2))
                        
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_rgb_dark); plt.title("Lower Contrast")
plt.subplot(132); plt.imshow(img_rgb); plt.title("Original")
plt.subplot(133); plt.imshow(img_rgb_bright); plt.title("Higher Contrast")
```

```python
# Handling Owerflow using np.clip

matrix1 = np.ones(img_rgb.shape)*.8
matrix2 = np.ones(img_rgb.shape)*1.2

img_rgb_dark = np.uint8(cv2.multiply(np.float64(img_rgb), matrix1))
img_rgb_bright = np.uint8(np.clip(cv2.multiply(np.float64(img_rgb), matrix2),0,255))
                        
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_rgb_dark); plt.title("Lower Contrast")
plt.subplot(132); plt.imshow(img_rgb); plt.title("Original")
plt.subplot(133); plt.imshow(img_rgb_bright); plt.title("Higher Contrast")
```
